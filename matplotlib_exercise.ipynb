```python
# 使うライブラリ
# !pip install japanize-matplotlib

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import japanize_matplotlib
japanize_matplotlib.japanize()
```

## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df = pd.read_csv('sample_data/hist_data.csv')
plt.yticks(np.arange(0,  30, 5))
plt.hist(df['aデータ'], bins=30, alpha=0.5, label="aデータ")
plt.hist(df['bデータ'], bins =30, alpha=0.5, label="bデータ")
plt.xlabel("温度(℃)")
plt.ylabel("度数")
plt.title("aデータとbデータの温度に関する度数分布")
plt.legend(loc="upper right")
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df =pd.read_csv("sample_data/scatter_data.csv")
d0=df[df["class"]==0]
d1=df[df["class"]==1]
d2=df[df["class"]==2]
plt.yticks(np.arange(0,  5, 0.5))
plt.scatter(d0['length (cm)'], d0['width (cm)'], c="r", label="class0")
plt.scatter(d1['length (cm)'], d1['width (cm)'], c="b", label="class1")
plt.scatter(d2['length (cm)'], d2['width (cm)'], c="g", label="class2")
plt.grid(True)
plt.title("クラス0～2のlengthとwidthに関する散布図")
plt.xlabel("length (cm)")
plt.ylabel("width (cm)")
plt.legend(loc="upper right")
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df=pd.read_csv("sample_data/line_data.csv")
x=df["期間"]
y=df["件数"]
plt.plot(x, y, marker="o")
plt.grid(True)
plt.title("期間ごとの件数の推移")
plt.xlabel("期間")
plt.ylabel("件数")
plt.xticks(rotation=25)
plt.show()

plt.xlabel("期間")
plt.ylabel("費用（円）")
a=df["期間"]
b=df["費用"]
plt.bar(a, b )
plt.grid(True)
plt.title("期間ごとの費用の推移")
plt.xticks(rotation=24)
plt.show()
plt.show()
```
